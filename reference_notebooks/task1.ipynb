## for-loop

### Get the square of every number

```python
X = [1, 2, 3, 4]

new_X = []
for x in X:
    new_X.append(x ** 2)
print(new_X)
```

### Get only the odd numbers

```python
X = [1, 2, 3, 4]

new_X = []
for x in X:
    is_odd = ((x % 2) == 0)
    if is_odd:
        new_X.append(x)
print(new_X)
```

### Get the square of the odd numbers

```python
X = [1, 2, 3, 4, 5, 6, 7]

new_X = []
for x in X:
    is_odd = ((x % 2) == 0)
    if is_odd:
        new_X.append(x ** 2)
print(new_X)
```

## List comprehension

### Get the square of every number

```python
X = [1, 2, 3, 4, 5, 6]
new_X = [x ** 2 for x in X]
print(new_X)
```

### Get the square of the odd numbers

```python
X = [1, 2, 3, 4, 5, 6]
new_X = [x ** 2 for x in X if x % 2 == 0]
print(new_X)
```

```python
X = [1, 2, 3, 4, 5, 6]
new_X = [x ** 2 
         for x in X 
         if x % 2 == 0]
print(new_X)
```

## Dictionary comprehension

```python
X = [1, 2, 3, 4, 5, 6]
new_X = {x: x ** 2 
         for x in X 
         if x % 2 == 0}
print(new_X)
```

```python
print(new_X.keys())
```
