```python
def function(param_1, param_2, param_3):
    var_1 = param_1 - param_2
    var_2 = param_3 & param_2
    var_3 = var_2 | var_1
    return var_3
```

```python
a = {1, 2, 3}
b = {3, 4, 5}
c = {5, 6, 7}
function(a, b, c)
```

```python
def function(param_1: set, 
             param_2: set, 
             param_3: set) -> set:
    var_1 = param_1 - param_2
    var_2 = param_3 & param_2
    var_3 = var_2 | var_1
    return var_3
```
